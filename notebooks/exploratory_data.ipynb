```python
import os
os.chdir('/home/bistiana/house_price_regression')
import pandas as pd

df = pd.read_csv('data/DATA-RUMAH.csv')
df
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_new = df.drop(['NO', 'NAMA RUMAH'], axis=True)
corr = df_new.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(11, 9))
sns.heatmap(corr, mask=mask, cmap='RdYlBu', vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

plt.title('Heatmap Korelasi')
plt.show()
```
